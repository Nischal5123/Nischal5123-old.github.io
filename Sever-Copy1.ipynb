```python
#import socket module
from socket import *
serverSocket = socket(AF_INET, SOCK_STREAM)
carriage_return="\r\n"
#Prepare a sever socket
port_number=6543
serverSocket.bind(('',port_number))#binding the socket to port number 80
serverSocket.listen(1)#max 1 connection per time
while True:
 #Establish the connection
 print ('Ready to serve...')
 connectionSocket, addr = serverSocket.accept()
 try:
     message = connectionSocket.recv(1024)
     filename = message.split()[1]
     f = open(filename[1:]) # starting from 1 to avoid the backslash
     outputdata = f.read()
     print (outputdata)
     #Send one HTTP header line into socket
     success_message="HTTP/1.1 200 OK\r\n\r\n"
     encoded_success_message=success_message.encode() 
     connectionSocket.send(encoded_success_message)
     #Send the content of the requested file to the client
     for i in range(0, len(outputdata)):
         output_data=outputdata[i]
         connectionSocket.send(output_data.encode()) #output data send one by one 
     connectionSocket.send(carriage_return.encode()) #carriage return and new line 
     connectionSocket.close()
 except IOError:
 #Send response message for file not found
   failure_message="HTTP/1.1 404 Not Found\r\n\r\n"
   encoded_failure_message=failure_message.encode() 
   connectionSocket.send(encoded_failure_message)
   html_message="<html><head></head><body><center><h1>404 Not Found</h1></center></body></html>\r\n"
   connectionSocket.send(html_message.encode())
 #Close client socket
   connectionSocket.close()

serverSocket.close() 
```
